```python
# Reshape the variables into NumPy arrays and put them into separate objects.
X = df['Cocoa_Pct'].values.reshape(-1,1)
y = df['Rating'].values.reshape(-1,1)
```

```python
#7 Split the data into two sets: a training set and a test set
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Create a regression object.

regression = LinearRegression()
```

```python

regression.fit(X_train, y_train)
```

```python

y_predicted = regression.predict(X_test)
```

```python

#9 Create a plot that shows the regression line on the test set.
```

```python
plot_test = plt
plot_test.scatter(X_test, y_test, color='brown', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Cocoa_Pct vs Rating (Test set)')
plot_test.xlabel('Cocoa_Pct')
plot_test.ylabel('Rating')
plot_test.show()
```

```python
#10 Write your own interpretation of how well the line appears to fit the data in a markdown cell.
#the linear model represented by the red regression line doesnâ€™t perfectly cover all of data points
```

```python
#11 Check the model performance statisticsâ€”MSE and R2 score.
```

```python
rmse = mean_squared_error(y_test, y_predicted) # This is the mean squared error
r2 = r2_score(y_test, y_predicted) # This is the R2 score. 
```

```python
print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```
